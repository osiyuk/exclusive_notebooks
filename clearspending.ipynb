```python
def dict_update(target, source):
    for key, value in source.items():
        
        if dict == type(value):
            target.setdefault(key, {})
            target[key] = dict_update(target[key], value)
            
        elif int == type(value):
            target.setdefault(key, 0)
            target[key] += value
            
        else:
            raise Exception("wtf?!" + str(value))

    return target
```

```python
def print_dict(data, lvl=0, indent=4):
    indent *= lvl
    indentation = ' ' * indent
    
    for key, value in data.items():
        print indentation + str(key) + ':',
        
        if dict == type(value):
            print ''
            print_dict(value, lvl + 1)
            
        elif list == type(value):
            print_list(value, lvl + 1)
            
        elif unicode == type(value):
            print value.encode('utf-8')
            
        else:
            print str(value)
            
def print_list(data, lvl=0, indent=4):
    indent_lvl = indent * (lvl - 1)
    indentation = (' ' * indent_lvl) + '-' + (' ' * (indent - 1))
    print '['
    for i, value in enumerate(data):
        if dict == type(value):
            print indentation
            print_dict(value, lvl)
            
        else:
            print indentation + str(value)
    
    print ' ' * indent_lvl + ']'
```
